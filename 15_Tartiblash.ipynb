! `DataFrame`ni qatorlari yoki ustunlari bo'yicha tartilash mumkin

```python
# KEling keyingi mashqlar uchun Githubdan kattaroq DF yuklab olamiz
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/usa_cars.csv', index_col=0)
df.head()
```

```python
df.shape
```

```python
df.sort_index(axis=1) # ustunlari bo'yicha tartiblash
```

```python
df.sort_index(axis=1, inplace=True) # o'zgarishlar joyida saqlash
df
```

```python
df.sort_index(ascending=False) # indeksi kamayish tartibida
```

```python
df.sort_index(axis=1, ascending=False) # ustuni kamayish tartibida
```

```python
df.sort_values(by='price')
```

```python
df.sort_values(by='price', ascending=False)
```

```python
df.sort_values(by=['price','mileage'], ascending=False)
```

```python
df.sort_values(by='year')
```
