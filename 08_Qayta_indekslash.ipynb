```python
import pandas as pd
import numpy as np
from pandas import DataFrame
```

```python
data = {'Yil':[2021, 2020, 2019, 2018, 2017, 2016, 2015, 2010],
        'Aholi soni':[33.9, 33.5, 32.9, 32.5, 31.9, 31.4, 30.9, 28.5],
        'Temp':[1.54, 1.48, 1.56, 1.62, 1.65, 1.66, 1.64, 1.53],
        'Zichlik':[79.77, 78.68, 77.53, 76.34, 75.13, 73.91, 72.71, 67.03]
}
df = pd.DataFrame(data)
df
```

```python
df.index = [21, 20, 19, 18, 17, 16, 15, 10] # indekslarni nomladik
df
```

```python
# Reindex
ind = [21, 20, 19, 18, 17, 16, 15, 10] # shablon uchun ro'yxat yaratdik
ind.sort() # tartiblab olamiz
ind
```

```python
df1 = df.reindex(ind) # qayta indekslangan yangi df hosil qilindi, asli o'zgarmagan holda
df1
```

```python
df1.reindex([10, 15, 20]) # tanlangan indekslar
```

```python
df1.reindex(index=range(10,22)) # oraliq indekslar qiymatini NaNga to'ldirdi
```

```python
df1.reindex(index=range(10,22), method='nearest') # taxminiy to'ldirish
```

Yuqoridagi nearest usuli har doim ham ma'qullanmaydi, ammo, masalan, dollar kursini taxmin qilish kabi misollarga mos kelishi mumkin

```python
df1.reindex(index=range(10,22), method='ffill')
```
