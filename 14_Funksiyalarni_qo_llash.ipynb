```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randn(3,4), index=list('abc'), columns=['Olma','Anor','Qovun','Anjir'])
df
```

```python
np.abs(df) # absolyut qiymat
```

```python
np.round(df) # yaxlitlash
```

```python
df.mean() # o'rtacha qiymat
```

```python
df.mean(axis=1)
```

```python
df.apply(np.abs) #df ga f-yani qo'llsh
```

```python
df.apply(sum)
```

```python
df.apply(sum, axis=1)
```

```python
# o'zimiz f-ya yozishimiz mumkin
f = lambda x: x.max() - x.min()
```

```python
df.apply(f) # df.apply(f = lambda x: x.max() - x.min()) desayam bo'ladi
```

```python
def foo(x):
    return pd.Series([x.min(), x.max()], index=['min', 'max'])
```

```python
df.apply(foo)
```
