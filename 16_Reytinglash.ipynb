```python
# KEling keyingi mashqlar uchun Githubdan kattaroq DF yuklab olamiz
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/usa_cars.csv', index_col=0)
df.head()
```

```python
df.sort_index(axis=1, inplace=True) # o'zgarishlar joyida saqlash
df
```

```python
df['rating'] = df['year'].rank() # qaysi yildagi kam bo'lsa reytingi baland
```

```python
df.sort_values('year')
```

```python
df['rating'] = df['price'].rank(method='min') # qancha arzon bo'lsa reyting baland
```
