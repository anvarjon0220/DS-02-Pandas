```python
import pandas as pd
import numpy as np
from pandas import DataFrame
```

```python
data = {'Yil':[2021, 2020, 2019, 2018, 2017, 2016, 2015, 2010],
        'Aholi soni':[33.9, 33.5, 32.9, 32.5, 31.9, 31.4, 30.9, 28.5],
        'Temp':[1.54, 1.48, 1.56, 1.62, 1.65, 1.66, 1.64, 1.53],
        'Zichlik':[79.77, 78.68, 77.53, 76.34, 75.13, 73.91, 72.71, 67.03]
}
df = pd.DataFrame(data, index=[21, 20, 19, 18, 17, 16, 15, 10])
df
```

```python
# Reindex
ind = [21, 20, 19, 18, 17, 16, 15, 10] # shablon uchun ro'yxat yaratdik
ind.sort() # tartiblab olamiz
ind
```

```python
df = df.reindex(ind) # qayta indekslangan yangi df hosil qilindi, asli o'zgarmagan holda
df
```

```python
df1 = df.drop(10) # 10-indeksni tashlash 
df1
```

```python
df1.drop([15,20]) # axis=0 standart qiymat berilgan, bir nechta indexni tashlash
```

```python
df1.drop('Yil', axis=1) # axis='columns' ustunni tashlash
```
